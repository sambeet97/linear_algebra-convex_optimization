```python
import numpy as np
```

```python
def steepest_descent(A, b):
    m,n = A.shape
    if m!= n:
        b = A.T@b
        A = A.T@A
    x = np.zeros(A.shape[0])
    print(x)
    for i in range(10000):

        r = b - A@x
        up = (np.inner(r,r))
        down = (np.inner(r,A@r))
        # up = r.T@r
        # down = r.T@A@r
        print(up)
        alpha = up/down
        x_temp =x 
        x = x + alpha * r

        if abs((x-x_temp).all()) < 0.001:
            print(i)
            break

    return x
```

```python
A = np.array([[1,2],[2,2], [3,4]])
b = np.array([1,2,3])
```

```python
steepest_descent(A, b)
```
